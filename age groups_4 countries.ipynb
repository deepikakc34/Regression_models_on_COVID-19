# Association of death rate with different age groups

# south korea

```python
ind_agegrp = agegrp_df[agegrp_df['country'] == 'south korea']
age = pd.get_dummies(ind_agegrp['age groups'])
rate = ind_agegrp['fatality rate']
age
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
model = regr.fit(age, rate)

print(f'the correlation value is {model.score(age, rate)}')
```
